```python
html = requests.get(
    "https://krisha.kz/a/show/1010380882",
    headers={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"}
).text
```

```python
soup = BeautifulSoup(html, 'html.parser')
script_tag = soup.find('script', {'id': 'jsdata'})
```

```python
print(script_tag is None)        # is the tag even found?
print(script_tag.string[:500])   # what does it actually contain?
```

```python
match = re.search(r'window\.data\s*=\s*(\{.+\});', script_tag.string, re.DOTALL)
data = json.loads(match.group(1))
print(data['advert'].get('lat'), data['advert'].get('lon'))
```

```python
matches = re.findall(r'51\.\d+', script_tag.string)
print(matches)
```
